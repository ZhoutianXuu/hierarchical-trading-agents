```python
import pandas as pd

def to_long_inplace(d: pd.DataFrame) -> pd.DataFrame:
    d = d.copy()
    if isinstance(d.columns, pd.MultiIndex) and d.columns.nlevels >= 2:
        names = list(d.columns.names or [])
        lvl = names.index('Ticker') if 'Ticker' in names else 1
        out = d.stack(level=lvl).reset_index()
        rename_map = {}
        if 'Date' in out.columns: rename_map['Date'] = 'date'
        if 'index' in out.columns: rename_map['index'] = 'date'
        if 'Ticker' in out.columns: rename_map['Ticker'] = 'ticker'
        if 'level_1' in out.columns: rename_map['level_1'] = 'ticker'
        out = out.rename(columns=rename_map)
        out.columns.name = None
        return out
    else:
        out = d.reset_index().rename(columns={'index':'date'})
        return out

for i in range(1, 21):
    name = f"df{i}"
    if name in globals():
        globals()[name] = to_long_inplace(globals()[name])
        try:
            print(name, "->", globals()[name].shape, "| cols:", list(globals()[name].columns[:6]), "...")
        except Exception:
            print(name, "done")
    else:
        print(f"{name} No exist")
```

```python
import re
import pandas as pd

for i in range(1, 21):
    name = f"df{i}"
    if name in globals():
        d = globals()[name]
        if isinstance(d, pd.DataFrame) and 'ticker' in d.columns:
            d = d.copy()
            d['ticker'] = d['ticker'].astype(str).str.replace(r'\.O$', '', regex=True) + '.O'
            globals()[name] = d
            print(name, 'done:', d['ticker'].unique()[:5])
        else:
            print(name, 'No')
    else:
        print(name, 'No')

```

```python
dfs = [globals()[f"df{i}"] for i in range(1, 21)]


combined = pd.concat(dfs, ignore_index=True, sort=False)
```
